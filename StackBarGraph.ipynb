```python
import numpy as np
import matplotlib.pyplot as plt


days = ['M', 'T', 'W', 'Th', 'F']
A = np.array([5, 9, 21, 2, 6])
B = np.array([7, 6, 4, 1, 2])
C = np.array([13, 17, 2, 19, 5])
D = np.array([11, 4, 5, 15, 5])
E = np.array([2, 12, 2, 15, 7])

ind = [x for x, _ in enumerate(days)]

plt.bar(ind, A, width=0.5, label='A', edgecolor = 'black', color='blue')
plt.bar(ind, B, width=0.5, label='B', edgecolor = 'black', color='cyan', bottom=A)
plt.bar(ind, C, width=0.5, label='C', edgecolor = 'black', color='grey', bottom=B+A)
plt.bar(ind, D, width=0.5, label='D', edgecolor = 'black', color='salmon', bottom=C+B+A)
plt.bar(ind, E, width=0.5, label='E', edgecolor = 'black', color='crimson', bottom=D+C+B+A)

plt.xticks(ind, days)
plt.yticks(np.arange(0, 61, 10))

plt.legend(loc="upper right")
plt.title("User Activity")
plt.grid(linestyle = 'dotted');

plt.show()
```
